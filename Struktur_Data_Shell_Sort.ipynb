```python
# Shell Sort In Python

def shellsort(data, n):

    # Rearrange elements at each n/2, n/4, n/8, ... intervals
    interval = n // 2
    while interval > 0:
        for i in range(interval, n):
            temp = data[i]
            j = i
            while j >= interval and data[j - interval] > temp:
                data[j] = data[j - interval]
                j -= interval

            data[j] = temp
        interval //= 2


size = len(data)
```

```python
shellsort(data, size)
print(data)
```
