## "Dumb" Bubble sort



````python
data=[2,5,7,11,2,1,4,8,12,3,7]
n=len(data)
````

````python
for j in range(n):
  for i in range(n-1):
    if data[i]>data[i+1]:
      temp = data[i]
      data[i] = data[i+1]
      data[i+1] = temp
    updateGraph(i,data)
````

## Bubble sort


````python
swap=True

for j in range(n):
  if (swap):
    swap=False
    for i in range(n-1):
      if data[i]>data[i+1]:
        temp = data[i]
        data[i] = data[i+1]
        data[i+1] = temp
        swap=True
      updateGraph(i,data)
````
